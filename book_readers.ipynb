```python
import pandas as pd
import numpy as np

df = pd.read_csv('books_db.csv')
authors = pd.read_csv('authors.csv')
language = pd.read_csv('language.csv')
publisher = pd.read_csv('publisher.csv')

publisher
```

```python
df = pd.merge(df, authors, left_on = 'author_id', right_index = True)
df = pd.merge(df, language, left_on = 'lang_id', right_index = True)
df = pd.merge(df, publisher, left_on = 'publisher_id', right_index = True)

df.head()
```

        در درسنامه‌های بعدی نحوه‌ی کار با این دستور را یاد خواهید گرفت.

```python
df.drop(columns = ['author_id', 'lang_id', 'publisher_id'], inplace = True)
df
```

```python
df.columns = ['bookID', 'title', 'average_rating', 'isbn', 'isbn13', 'num_pages', 'ratings_count', 'text_reviews_count', 'publication_date', 'author', 'lang', 'publisher']
df.head()
```

```python
grouped = df.groupby('publisher')
publisher_rating = grouped['average_rating'].agg('mean')
publisher_rating
```

```python
conditions = [ (df['num_pages'] < 250), ((df['num_pages'] >= 250) & (df['num_pages'] <= 499)), ((df['num_pages'] >= 500) & (df['num_pages'] <= 749)), ((df['num_pages'] >= 750) & (df['num_pages'] <= 999)), (df['num_pages'] >= 1000) ]
choices = ['<250', '250-499', '500-749', '750-999', '>=1000']
df['num_pages_grp'] = np.select(conditions, choices)
df
```

```python
author_page = df.groupby(['author','num_pages_grp'])['num_pages_grp'].size()
author_page
```
